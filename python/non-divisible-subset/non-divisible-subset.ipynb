# The Solution

```python
from collections import defaultdict

def remainders(k, S):
    remains = defaultdict(int)
    for s in S:
        remainder = s % k
        remains[remainder] += 1
    return remains
    
remainders(4, [19, 10, 12, 10, 24, 25, 22])
```

```python
def nonDivisibleSubset(k, S):
    rs = remainders(k, S)
    count = 0
    for i in range(0, k//2+1):
        j = k - i
        if i == 0 or i == j:
            if rs[i] or rs[j]:
                count += 1
        else:
            count += max(rs[i], rs[j])
    return count

nonDivisibleSubset(4, [19, 10, 12, 10, 24, 25, 22])
```
